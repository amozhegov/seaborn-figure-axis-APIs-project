```python
import seaborn as sns
```

# Figure and Axis API

```python
penguins = sns.load_dataset('penguins')
```

```python
type(penguins)
```

```python
penguins.head(5)
```

```python
penguins.tail(5)
```

```python
# Body mass and species correlation
ax = sns.scatterplot(data=penguins, x='body_mass_g', y='species')
```

```python
type(ax)
```

```python
ax = sns.relplot(data=penguins, x='body_mass_g', y='species', kind='scatter')
```

```python
ax = sns.relplot(data=penguins, x='body_mass_g', y='species', kind='scatter', col='sex', hue='island')
```

```python
ax = sns.relplot(data=penguins, x='body_mass_g', y='species', kind='scatter', hue='sex')
```

```python
ax = sns.relplot(data=penguins, x='body_mass_g', y='species', kind='scatter', size='sex')
```

```python
sns.displot(data=penguins, x='body_mass_g', kind='kde', hue='species', fill=True)
```

```python
flights = sns.load_dataset('flights')
flights.head(10)
```

```python
sns.relplot(data=flights, x='year', y='passengers', kind='line')
```

# Distribution Plots

```python
sns.displot(data=flights, x='passengers', y='year', kind='hist', hue='month')
```

```python
# JointPlot
sns.jointplot(data=penguins, x='bill_length_mm', y="flipper_length_mm")
```

```python
# JointPlot
sns.jointplot(data=penguins, x='bill_length_mm', y="flipper_length_mm", hue='species')
```

```python
# PairPlot
sns.pairplot(data=penguins, hue='species')
```

```python
sns.catplot(data=penguins, x='species', y='flipper_length_mm')
```

```python
sns.catplot(data=penguins, x='species', y='flipper_length_mm', kind='box')
```

# Regression Visualization

```python
# By sex
sns.lmplot(data=penguins, x='bill_length_mm', y="flipper_length_mm", hue='sex')
```

```python
# By island
sns.lmplot(data=penguins, x='bill_length_mm', y="flipper_length_mm", hue='island')
```

# Vizual Style Change

```python
sns.set_style('darkgrid')
sns.lmplot(data=penguins, x='bill_length_mm', y="flipper_length_mm", hue='island')
```
